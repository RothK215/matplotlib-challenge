```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import numpy as np
%matplotlib inline
# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single dataset
mouse_tumors = pd.merge(mouse_metadata, study_results)

# Display the data table for preview
mouse_tumors = mouse_tumors[['Mouse ID', 'Timepoint', 'Tumor Volume (mm3)', 'Metastatic Sites', 'Drug Regimen', 'Sex', 'Age_months', 'Weight (g)']]
mouse_tumors
```

```python
# Checking the number of mice.
mice_count = len(mouse_tumors['Mouse ID'].unique())
mice_count
```

```python
# Getting the duplicate mice by ID number that shows up for Mouse ID and Timepoint. 
duplicates = mouse_tumors[mouse_tumors.duplicated(['Mouse ID', 'Timepoint'])]
duplicates
```

```python
# Optional: Get all the data for the duplicate mouse ID.
double_info = mouse_tumors.loc[mouse_tumors['Mouse ID']=='g989']
double_info
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
clean_df = mouse_tumors[mouse_tumors['Mouse ID'] !='g989']
clean_df
```

```python
# Checking the number of mice in the clean DataFrame.
mice_count2 = len(clean_df['Mouse ID'].unique())
mice_count2
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen
# Use groupby and summary statistical methods to calculate the following properties of each drug regimen: 
# mean, median, variance, standard deviation, and SEM of the tumor volume. 

tumor_ave = clean_df.groupby(['Drug Regimen']).mean()['Tumor Volume (mm3)']
tumor_med = clean_df.groupby(['Drug Regimen']).median()['Tumor Volume (mm3)']
tumor_var = clean_df.groupby(['Drug Regimen']).var()['Tumor Volume (mm3)']
tumor_std = clean_df.groupby(['Drug Regimen']).std()['Tumor Volume (mm3)']
tumor_err = clean_df.groupby(['Drug Regimen']).sem()['Tumor Volume (mm3)']
# Assemble the resulting series into a single summary DataFrame.

drug_stats = pd.DataFrame({"Mean Tumor Volume": tumor_ave,
                           "Median Tumor Volume": tumor_med,
                           "Tumor Volume Variance": tumor_var,
                           "Tumor Volume Std. Dev": tumor_std,
                           "Tumor Volume Std. Err.": tumor_err
                          })
drug_stats
```

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen
# Using the aggregation method, produce the same summary statistics in a single line.
drug_agg = clean_df.groupby(['Drug Regimen']).agg(['mean', 'median', 'var', 'std', 'sem'])['Tumor Volume (mm3)']
drug_agg
```

## Quartiles, Outliers and Boxplots

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:  
    # Capomulin, Ramicane, Infubinol, and Ceftamin

# Start by getting the last (greatest) timepoint for each mouse.
#timepoint_max = clean_df[clean_df['Timepoint']>=45]
timepoint_max = clean_df.drop_duplicates(subset=['Mouse ID'], keep='last')
# Merge this group df with the original DataFrame to get the tumor volume at the last timepoint.
df_merge = pd.merge(mouse_tumors, timepoint_max)
df_merge
```

```python
# Put treatments into a list for for loop (and later for plot labels)
treatments = ['Capomulin', 'Ramicane', 'Infubinol', 'Ceftamin']
# Create empty list to fill with tumor vol data (for plotting)
tumor_vol = []

# Calculate the IQR and quantitatively determine if there are any potential outliers. 
    # Locate the rows which contain mice on each drug and get the tumor volumes
for t in treatments:
    
    drug_data = df_merge.loc[df_merge['Drug Regimen']==t, "Tumor Volume (mm3)"]
    
    # add subset 
    tumor_vol.append(drug_data)
    
    # Determine outliers using upper and lower bounds
    quartile = drug_data.quantile([0.25, 0.5, 0.75])
    lowerquart = quartile[0.25]
    upperquart = quartile[0.75]
    
    iqr = upperquart-lowerquart
    medianquart = quartile[0.5]
    
    lowerbound = lowerquart-(1.5*iqr)
    upperbound = upperquart+(1.5*iqr)
```

```python
# Generate a box plot that shows the distrubution of the tumor volume for each treatment group.
figure, ax1=plt.subplots()
x = [1, 2, 3, 4]
plt.xticks(x, treatments, rotation='horizontal')
ax1.set_ylabel("Final Tumor Volume (mm3)")

plt.boxplot(tumor_vol, treatments)
plt.show()
```
